```python
def confidence(row):
    if row == 'Полностью доверяю':
        a = 1
    if row == 'Скорее доверяю':
        a = 2
    if row == 'Скорее не доверяю':
        a = 3
    if row == 'Полностью не доверяю':
        a = 4
    if row == 'Затрудняюсь ответить':
        a = -1
    if row == 'Абсолютно доверяю':
        a = 1
    if row == 'Абсолютно не доверяю':
        a = 4
    return a
```

Имиграционная политика

1 - пускать любого, кто захочет

2 - пускать пока есть работа

3 - квоты

4 - никого не пускать

-1 - затрудняюсь

```python
df['По вашему мнению, какой должна быть иммиграционная политика?'].unique()
```

```python
def immigrant(row):
    if row == 'Мы должны пускать всех желающих':
        a = 1
    if row == 'Пускать всех желающих до тех пор, пока есть рабочие места':
        a = 2
    if row == 'Установить жесткие ограничения на количество иностранцев, которые могут сюда приехать':
        a = 3
    if row == 'Запретить приезжать сюда из других стран':
        a = 4
    if row == 'Затрудняюсь ответить':
        a = -1
    return a
```

```python
df['Q130: Immigration policy preference'] = df['По вашему мнению, какой должна быть иммиграционная политика?'].apply(immigrant)
df
```
